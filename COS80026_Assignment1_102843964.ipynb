```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('COS80026_Assignment1_creditcard.csv')
df.head()
```

```python
print('MIssing Values in columns: \n',df.isna().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
X = df.drop(columns = 'Class',axis = 1)
plt.figure(figsize=(15,15))
sns.heatmap(X.corr(),annot = True, fmt='.1f', vmin=-1, vmax=1, center= 0,cmap='coolwarm')
plt.show()
```

```python
df.groupby(by='Class').count().reset_index()[['Class','V1']].rename(columns = {'V1':'Count'})
```

```python
X = np.array(X)
y = np.array(df['Class']).reshape(-1,1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=201)
tree = DecisionTreeClassifier(random_state=201,class_weight= {0:1,1:100},ccp_alpha= 1e-4,
                              max_leaf_nodes=120,min_impurity_decrease=1e-3,max_features=10)
tree.fit(X_train,y_train)
ypreds = tree.predict(X_test)
probs = tree.predict_proba(X_test)
plot_confusion_matrix(tree,X_test, y_test)
plt.show()
print('Accuracy Score in test set: ',accuracy_score(y_test,ypreds))
fpr, tpr, thresholds = roc_curve(y_test, probs[:,1], pos_label=1)
plt.plot(fpr,tpr)
plt.grid()
plt.title('AUC under ROC of postive class (class 1):'+str(auc(fpr,tpr)))
plt.show()
```
